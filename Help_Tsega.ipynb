```python
def make(excel_file):
    x1 = pd.read_excel(excel_file,skiprows=12,skipfooter=17)[['Monitored Developmental Milestone','GMP/ Nutrition','EPI','U5 clinic/ IMNCI','Total']]
    x2 = pd.read_excel(excel_file,skiprows=18,skipfooter=14)[['Counseling/key message on child development at maternal service touch points','ANC','Delivery','PNC','Total']]
    x3 = pd.read_excel(excel_file,skiprows=21,skipfooter=10).iloc[:,1:6]
    x4 = pd.read_excel(excel_file,skiprows=25,skipfooter=8)[['HOME VISIT/COMMUNITY','All HHs','Vulnerable* HHs']]
    x5 = pd.read_excel(excel_file,skiprows=27,skipfooter=6).iloc[:,1:5]
    x6 = pd.read_excel(excel_file,skiprows=29,skipfooter=2,header=None,names=['fields','Total','','','','']).iloc[:,0:2]
    return [x1,x2,x3,x4,x5,x6]

def united(files,fields='pass',value='pass'):
    if fields or value == 'pass':
        checked = files
        for file in files:
            info=pd.read_excel(file,skiprows=3,skipfooter=25,index_col=('Fields'))['value'].to_frame()
            print(info)
    else:
        checked=[]
        for file in files:
            info=pd.read_excel(file,skiprows=3,skipfooter=25,index_col=('Fields'))['value'].to_frame()
            if info.loc[fields.upper(),'value']==value:
                checked.append(file)
                print(file+'\n\n')
                print(info)


    dfs=[]
    for file in checked:
        dfs.append(make(file))
    base = make('/content/ECD Monthly report form.xlsx')
    for df in dfs:
        for i in range(0,6):
            base[i] = base[i].add(df[i],fill_value=0)
    return base
print('done!')
```
